```python
# Import Packages
import pandas as pd
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle
```

```python
df = pd.read_csv(r"C:\Users\zacha\龙京\python\youtube\uci_student\student-por.csv", sep = ";")
```

```python
df.head()
```

```python
# Cutting Columns
df = df[["G1", "G2", "G3", "studytime", "failures", "absences"]]
```

```python
# Assigning Target Variable
predict = ["G3"]
```

```python
# Establishing X and Y Values
import numpy as np 
X = np.array(df.drop(predict, axis = 1))
y = np.array(df[predict])
```

```python
# Assigning Train and Test
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size = 0.1)
```

```python
# Testing Model Accuracy
linear = linear_model.LinearRegression()

linear.fit(x_train, y_train)
accuracy = linear.score(x_test, y_test)
print(accuracy)
```

```python
# Coefficient and Intercept
print('Co: \n', linear.coef_)
print('Intercerpt: \n', linear.intercept_)
```

```python
# Model Prediction
predictions = linear.predict(x_test)
```

```python
for x in range(len(predictions)):
    print(predictions[x], x_test[x], y_test[x])
```
