```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

# Line Plot

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

x2 = np.arange(0,4.5,0.5)

plt.figure(figsize = (7,5))

#plt.plot(x,y, label = '2x', color = 'red', linewidth = 2, marker = '.', markersize = 10, markeredgecolor = 'yellow')

plt.plot(x,y,'b.--', label = '2x')
plt.plot(x2,x2**2, 'r--o', label = 'x^2')
plt.plot(x2[:6],x2[:6]**2, 'g-o', label = 'x2[:6]')

plt.legend(loc = 2, fontsize = 'x-small')
plt.title('First Graph', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 18})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10,12,14,16])

plt.grid(True)
plt.tight_layout()
plt.style.use('fivethirtyeight')
plt.show()
```

# Bar Chart 

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values, color ='b')
    
bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')

plt.show()
```
