```python
plt.plot(gas.Year, gas.USA, 'bo-', linewidth = 2, label = 'USA')
plt.plot(gas.Year, gas.Canada, 'go-', linewidth = 2, label = 'Canada')
plt.plot(gas.Year, gas['South Korea'], 'ro-', linewidth = 2,label = 'South Korea')

plt.legend(loc = 2, fontsize = 'xx-small')
plt.title('Gas Prices (USD)', fontdict = {'fontweight': 'bold', 'fontsize': 18})
plt.xlabel('Years')
plt.ylabel('Price (USD)')

plt.xticks(gas.Year[::3])

plt.style.use('fivethirtyeight')
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv(r"C:\Users\zacha\Hyperborea\csv files\fifa_data.csv")

left = df.loc[df['Preferred Foot'] == 'Left'].count()[0]
right = df.loc[df['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left', 'Right']

plt.pie([left, right], autopct = '%.2f%%', labels = labels)

plt.legend()
plt.title('FIFA Foot Preference')

plt.style.use('ggplot')
plt.show()
```

```python
plt.figure(figsize=(5,8), dpi=100)

barcelona = df.loc[df.Club == "FC Barcelona"]['Overall']
madrid = df.loc[df.Club == "Real Madrid"]['Overall']
revs = df.loc[df.Club == "New England Revolution"]['Overall']

bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    box.set(color='#4286f4', linewidth=2)
    box.set(facecolor = '#e0e0e0')
    
plt.style.use('default')
plt.show()
```
