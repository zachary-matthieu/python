```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y, s = 100, c = 'yellow', marker = 'o', edgecolor = 'black', linewidth = 2, alpha = 0.75)

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.style.use('seaborn')

plt.show()
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]

plt.scatter(x, y, s = sizes, c = colors, cmap = 'Greens', marker = 'o', edgecolor = 'black', linewidth = 2, alpha = 0.75)

# Color Bar 

cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.style.use('seaborn')

plt.show()
```

```python
df = pd.read_csv(r'C:\Users\zacha\Hyperborea\csv files\views_data.csv')
view_count = df['view_count']
likes = df['likes']
ratio = df['ratio']

plt.scatter(view_count, likes, c = ratio, cmap = 'summer', marker = 'o', edgecolor = 'black', linewidth = 2, alpha = 0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.style.use('seaborn')

plt.show()
```
