```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\zacha\Hyperborea\csv files\dev_salary.csv')

ages = df['Age']
dev_salaries = df['All_Devs']
py_salaries = df['Python']
js_salaries = df['JavaScript']

plt.plot(ages, dev_salaries, color = '#444444', linestyle = '--', label = 'All Devs')

plt.plot(ages, py_salaries, label = 'Python')


# Median Fill

overall_mean = 69257

# Applying Conditional Fills

plt.fill_between(ages, py_salaries, overall_mean, where = (py_salaries > overall_mean), interpolate = True, color = 'green', alpha = 0.25)

# Contrasting Fills

plt.fill_between(ages, py_salaries, overall_mean, where = (py_salaries <= overall_mean), interpolate = True, color = 'red', alpha = 0.25)


plt.legend(loc = 2, fontsize = 'x-small')

plt.title('Developer Salaries')
plt.xlabel('Ages')
plt.ylabel('Salary(USD)')

plt.tight_layout()

plt.grid(True)

plt.style.use('seaborn-pastel')

plt.show()
```
